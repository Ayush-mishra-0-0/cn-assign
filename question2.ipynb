# Graph Centrality Measures ðŸš€

This notebook calculates **Degree**, **Closeness**, and **Betweenness Centrality** for various graphs using NetworkX. The measures are defined as follows:

- **Degree Centrality**:  
  ![equation](https://latex.codecogs.com/png.latex?C_D(v)=\frac{deg(v)}{N-1})  
  Normalized by (N-1).

- **Closeness Centrality**:  
  ![equation](https://latex.codecogs.com/png.latex?C_C(v)=\frac{N-1}{\sum{d(v,u)}})  
  It is set to 0 if the graph is disconnected.

- **Betweenness Centrality**:  
  For every pair (s, t), compute all shortest paths and for every intermediate node, add the fraction of paths that pass through it. Finally, normalize by  
  ![equation](https://latex.codecogs.com/png.latex?\frac{2}{(N-1)(N-2)}).

```python
!pip install networkx

import networkx as nx
import matplotlib.pyplot as plt

def my_degree_centrality(g):
    centrality = {}
    n = len(g.nodes())
    for node in g.nodes():
        centrality[node] = g.degree(node) / (n - 1)
    return centrality

def my_closeness_centrality(g):
    centrality = {}
    n = len(g.nodes())
    for node in g.nodes():
        distances, queue = {node:0}, [node]
        while queue:
            cur = queue.pop(0)
            for nei in g.neighbors(cur):
                if nei not in distances:
                    distances[nei] = distances[cur] + 1
                    queue.append(nei)
        total = sum(distances.values())
        centrality[node] = (n - 1) / total if total and len(distances)==n else 0.0
    return centrality

def find_all_shortest_paths(g, source, target):
    if source == target: return [[source]]
    paths, queue, shortest = [], [[source]], None
    while queue:
        path = queue.pop(0)
        if shortest and len(path) > shortest: break
        for nei in g.neighbors(path[-1]):
            if nei in path: continue
            new_path = path + [nei]
            if nei == target:
                if shortest is None: shortest = len(new_path)
                if len(new_path)==shortest: paths.append(new_path)
            else:
                queue.append(new_path)
    return paths

def my_betweenness_centrality(g):
    centrality = {node:0.0 for node in g.nodes()}
    nodes = list(g.nodes())
    n = len(nodes)
    for i in range(n):
        for j in range(i+1, n):
            s, t = nodes[i], nodes[j]
            paths = find_all_shortest_paths(g, s, t)
            if not paths: continue
            for v in g.nodes():
                if v==s or v==t: continue
                centrality[v] += sum(1 for path in paths if v in path[1:-1]) / len(paths)
    norm = 2/((n-1)*(n-2))
    for node in centrality: centrality[node] *= norm
    return centrality

graph_specs = [
    (range(1,7), [(1,2),(1,3),(2,3),(2,4),(4,6),(4,5),(5,6)]),
    (range(1,9), [(1,2),(2,3),(2,5),(2,4),(4,6),(5,6),(7,6),(6,8)]),
    (range(1,6), [(1,5),(2,5),(2,3),(2,4)]),
    (range(1,11), [(1,3),(1,5),(1,6),(2,3),(2,5),(3,4),(5,10),(5,8),(5,9),(10,7)]),
    (range(1,16), [(1,12),(1,2),(1,4),(4,5),(4,6),(5,9),(6,9),(6,8),(9,3),(9,10),(9,7),(9,11),(11,13),(13,15),(13,14)]),
    (range(1,15), [(8,14),(8,5),(5,2),(5,9),(5,1),(2,3),(2,10),(3,4),(3,13),(10,7),(13,6),(6,10),(6,1),(6,9),(6,11),(9,12),(1,10),(1,3),(9,11)]),
    (range(1,11), [(1,5),(1,9),(2,7),(2,10),(2,5),(3,9),(3,4),(4,5),(5,6),(5,7),(5,10),(6,9),(8,9)])
]

for nodes, edges in graph_specs:
    g = nx.Graph()
    g.add_nodes_from(nodes)
    g.add_edges_from(edges)
    print()
    print("-----Degree Centrality:-------")
    print(my_degree_centrality(g))
    print()
    print("-------Betweenness Centrality:----------")
    print(my_betweenness_centrality(g))
    print()
    print("-------Closeness Centrality:----------")
    print(my_closeness_centrality(g))
    print()
    pos = nx.spring_layout(g)
    nx.draw(g, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=15)
    el = {edge:'' for edge in g.edges()}
    nx.draw_networkx_edge_labels(g, pos, edge_labels=el, font_color='blue')
    plt.show()
```
